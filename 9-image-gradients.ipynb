Image Gradients
===

There are three types of gradient filters provided by OpenCV: (1) Sobel, (2) Scharr, and (3) Laplacian.

## Sobel and Scharr Derivatives

Sobel operators are a joint Gaussian smoothing and differentiation operation. Hence, it is more resistant to noise. We can specify the direction of the derivatives we will take, either horizontal or vertical. We can also specify the kernel size.

## Laplacian Derivatives

It calculates the Laplacian of a given image by the following relation $\Delta src = \dfrac{\partial^{2} src}{\partial x^{2}} + \dfrac{\partial^{2} src}{\partial y^{2}}$.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('figures/sudoku.JPG', 0)

laplacian = cv2.Laplacian(image, cv2.CV_64F)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

plt.figure(figsize=(20, 20))
plt.subplot(221), plt.imshow(image, cmap='gray')
plt.subplot(222), plt.imshow(laplacian, cmap='gray')
plt.subplot(223), plt.imshow(sobel_x, cmap='gray')
plt.subplot(224), plt.imshow(sobel_y, cmap='gray')

plt.show()
```
